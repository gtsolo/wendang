```python
import pandas as pd
import numpy as np
```

```python
dict1 = {'ID':[1,2,3,4,5,6,7,8,9],
         'System':['win10','win10',np.nan,'win10',
                np.nan,np.nan,'win7','win7','win8'],
      'cpu':['i7','i5',np.nan,'i7',np.nan,np.nan,'i5','i5','i3']}

dict2 = {'ID':[1,2,3,4,5,6,7,8,9],
         'System':[np.nan, np.nan,'win7',np.nan,
                'win8','win7',np.nan,np.nan,np.nan],
        'cpu':[np.nan,np.nan,'i3',np.nan,'i7',
                'i5',np.nan,np.nan,np.nan]}
df1=pd.DataFrame(dict1)
df2=pd.DataFrame(dict2)

print(df1)
print(df2)
print(pd.concat([df1,df2],axis=1))
df2['tmp']=4 
df2
```

```python
pd.concat([df1,df2],join='inner')
pd.concat([df1,df2],join='outer')
```

```python
df1.append(df2)  #纵向合并
```

检测与处理异常值

```python
import matplotlib.pyplot as plt

data=pd.read_excel("D:/大数据案例/data/meal_order_detail.xlsx")
data.head()
```

```python
plt.boxplot(data['amounts'])
plt.show()
```

```python
def replace(x):
    import numpy as np
    QU=x.quantile(0.75)
    QL=x.quantile(0.25)
    IQR=QU-QL
    x=x.copy()
    x[(x>(QU+1.5*IQR)) | (x<(QL-1.5*IQR))]=np.nan
    return x
```

```python
data['amounts'].isnull().sum()
```

```python
replace(data['amounts']).isnull().sum()
```

标准化数据

```python
import matplotlib.pyplot as plt

data=pd.read_excel("D:/大数据案例/data/meal_order_detail.xlsx")
data.head()
```

```python
data[['counts','amounts']].describe()
```

```python
#标准化数据

def min_max_scale(x):
    return (x-x.min())/(x.max()-x.min())

def stand_scale(x):
    return (x-x.mean())/x.std()

def decimal_scale(x):
    import numpy as np
    k = np.ceil(np.log10(x.abs().max()))
    return x/10**k
```

```python
data[['counts','amounts']].agg(min_max_scale)
data[['counts','amounts']].agg(stand_scale)
data[['counts','amounts']].agg(decimal_scale)
```
