#数据框的增删改查

```python
d=[[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df = pd.DataFrame(d,index=['a','b','c','d'],columns=['A','B','C','D'])
df['A']   #单列数据访问
print(df[['A','D']])  #多列数据访问
print(df.head(3))
print(df.tail(3))
print(df.iloc[0,0])  #按照行列顺序访问
print(df.iloc[0:3,1])
print(df.iloc[0,:])
print(df.iloc[1:3,1:3])

print(df.loc['a','A'])   #按照行列名称访问
print(df.loc['a':'c','A'])
print(df.loc[:,'A' ])
print(df.loc['a',:])
print(df.loc[['b','c'],['B','C']])


#注意以下访问返回值不同
print(df.iloc[:,0])
print(type(df.iloc[:,0]))
print(df.iloc[:,0:1])
print(type(df.iloc[:,0:1]))
```

#修改数据框元素

```python
d=[[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df = pd.DataFrame(d,index=['a','b','c','d'],columns=['A','B','C','D'])
print(df)
df.loc['a','A']=100
df.loc[:,'B']=0.2
df.loc[:,'C']=[1,2,3,4]
print(df)
```

```python
#增加数据
d=[[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df = pd.DataFrame(d,index=['a','b','c','d'],columns=['A','B','C','D'])
print(df)
df['E']=5
df['F']=[1,2,3,4]
print(df)
```

```python
#删除数据
d=[[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df = pd.DataFrame(d,index=['a','b','c','d'],columns=['A','B','C','D'])
print(df)
print(df.drop('D',axis=1))
print(df)

print(df.drop('D',axis=1,inplace=True))
print(df)

print(df.drop(['a','c'],axis=0))
```

#使用分组聚合进行组内计算

```python
import pandas as pd
detial = pd.read_excel('D:/大数据案例/data/meal_order_detail.xlsx')
detial.head()
```

```python
detial_group= detial[['order_id','counts','amounts']].groupby(by='order_id')
```

```python
detial_group
```

```python
detial_group.agg('count')
```

```python
detial_group.agg(['mean','sum'])
```

```python
detial_group.agg({'counts':['mean','max'],'amounts':'std'})
```

```python
detial_group.agg({'counts': lambda x:sum(x)**2,'amounts':'std'})
```

创建数据透视表

```python
import pandas as pd
detial = pd.read_excel('D:/大数据案例/data/meal_order_detail.xlsx')
detial.head()
```

```python
pd.pivot_table(detial[['order_id','counts','amounts']],index='order_id',aggfunc='sum')
```

```python
pd.pivot_table(detial[['order_id','dishes_name','counts']],index='order_id',columns='dishes_name',aggfunc=('sum')).head(3)
```

```python
pd.pivot_table(detial[['order_id','dishes_name','counts']],index='order_id',columns='dishes_name',values='counts',fill_value=0).head(3)
```

```python
pd.crosstab(index=detial['order_id'],columns=detial['dishes_name'],values=detial['counts'],aggfunc='sum').fillna(0)
```
