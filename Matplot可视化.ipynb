了解绘图和基本参数

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.arange(0,1.1,0.1)
print(x)
plt.figure()      #创建画布
plt.plot(x,x**2)  #第二环节，绘制图形
plt.plot(x,x**3)
plt.show()        #第三环节，显示图形
```

```python
x= np.arange(0,1.1,0.1)
print(x)
plt.figure()      #创建画布
plt.plot(x,x**2)  #第二环节，绘制图形
plt.xlim(0,1)    #x轴坐标
plt.ylim(0,1)    #y轴坐标
plt.rcParams['font.sans-serif']=['simhei']
plt.title('线性')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,x**3)
plt.legend(['x=y^2','x=y^3'])
plt.savefig('example.png')
plt.show()        #第三环节，显示图形
```

```python
import numpy as np
import matplotlib.pyplot as plt
data =np.load('D:\大数据案例\data\国民经济核算季度数据.npz',allow_pickle=True)
data.files
columns = data['columns']
values = data['values']
print(columns)
print(values)
data['values'].shape
```

绘制散点图

```python
plt.figure(figsize=(8,6))

plt.scatter(values[:,1],values[:,3],marker='o')
plt.scatter(values[:,1],values[:,4],marker='*')
plt.scatter(values[:,1],values[:,5],marker='d')

plt.xticks(range(0,70,4),values[range(0,70,4),1],rotation=45)
plt.ylabel('生产总值(亿元)',rotation=0)
plt.legend(['第一产业总值','第二产业总值','第三产业总值'])
plt.title('2010-2017年各产业生产总值')
plt.savefig('2010-2017年各产业生产总值.png')
plt.show()
```

绘制折线图

```python
plt.figure(figsize=(8,6))

plt.plot(values[:,1],values[:,3],linestyle='-.')
plt.plot(values[:,1],values[:,4],linestyle=':')
plt.plot(values[:,1],values[:,5],marker='d')

plt.xticks(range(0,70,4),values[range(0,70,4),1],rotation=45)
plt.ylabel('生产总值(亿元)',rotation=0)
plt.legend(['第一产业总值','第二产业总值','第三产业总值'])
plt.title('2010-2017年各产业生产总值')
plt.savefig('2010-2017年各产业生产总值.png')
plt.show()
```

绘制饼图

```python
plt.figure(figsize=(8,6))
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False

label=['第一产业总值','第二产业总值','第三产业总值']
plt.pie(values[-1,3:6],explode=[0.04,0.04,0.04],labels=label,autopct='%1.1f%%')
plt.title('2017年生产总值')
plt.show()
```

绘箱线图

```python
plt.figure(figsize=(8,6))
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False

label=['第一产业总值','第二产业总值','第三产业总值']
plt.boxplot(values[:,3:6],notch=True,labels=label)
plt.show()
```
