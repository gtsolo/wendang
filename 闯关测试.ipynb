```python
read =pd.read_excel('Data.xlsx')

w = read.loc[read['日期']>'2015-10-08']
wend=w['日期'].unique()
wend
wenday=read.loc[read['日期']>'2015-10-08']['日期'].unique()
print(wenday)
```

```python
  
for c in wenday:  
    d=read.loc[read['日期'].values == c]  
    df1.append(d)
```

```python
station = read.iloc[:,0].unique()  
time =read.iloc[:,2].unique()
A1 =[]  
A2 =[]  
A3 =[]  
A4 =[]  
for i in range(len(station)):  
    d1=w.iloc[w['站点编号'].values==station[i],:]  
    for j in range(len(time)):  
        sk = d1['时刻'].unique()  
        if time[j] in sk:  
            jz_sum = d1.iloc[d1['时刻'].values==time[j],3].sum()  
            cz_sum = d1.iloc[d1['时刻'].values==time[j],4].sum()  
            A1.append(station[i])  
            A2.append(time[j])  
            A3.append(jz_sum)  
            A4.append(cz_sum)  
df0=pd.DataFrame({'站点编号':A1,'时刻':A2,'总进站客流':A3,'总出站客流':A4})        
df0.to_excel('各站点各时刻进出站客流数据.xlsx') 
```

```python
Numbers=[i for i in range(1,31)]#Numbers课程设计编号  
while len(Numbers):  
    Start=input('是否进行抽签(yes or no):' )#确认运行程序的人是否需要进行抽签  
    if Start=='yes':  
        R_max=len(Numbers)#R_max，随机数的最大取值  
        rand_num=random.randint(0,R_max-1)#生成的随机数  
        Draw=Numbers[rand_num]#Draw随机抽取的课程编号  
        Numbers.remove(Draw)#删除每一次抽取的Draw  
        print("您抽取到的课程设计编号为：",Draw)  
    else:  
        print("程序已经结束")  
        break;
```
